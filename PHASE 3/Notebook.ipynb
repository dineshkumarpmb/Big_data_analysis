```python
                          #Data visualisation for year and avg temp
import pandas as pd
import matplotlib.pyplot as plt
import csv
import numpy as np

data = pd.read_csv('Global-temperature.csv')

data['year'].plot.hist()
data['AverageTemperatureFahr'].plot.hist()
plt.xlabel('year')
plt.ylabel('avg temp')
plt.title('GLOBAL TEMP')
plt.show()
```

```python
import matplotlib.pyplot as plt 
import csv 

x = [] 
y = [] 

with open('Global-temp.csv','r') as csvfile: 
	plots = csv.reader(csvfile, delimiter = ',') 
	
	for row in plots: 
		x.append(row[5]) 
		y.append(row[3])

plt.bar(x, y, color = 'r', label = "temperature  °C ") 
plt.xlabel('country') 
plt.ylabel('Avg Temp') 
plt.legend() 
plt.show() 
```
